1. Using breast cancer coimbre dataset Perform Data exploratory analysis on the data
2. Use 30% of data as the test set and build following models to predict Labels variable                      
 a) Build the Na√Øve Bayes model to predict Labels variable                       
 b) Build the Decision tree model to predict Labels variable

INPUT: Breast cancer coimbre data set

ASSUMPTIONS: None

SOLUTION DESIGN:
1. Divide the dataset into train, validation, test 
2. Handling missing values
3. Handling duplicates
4. Build Naive Bayes Model
5. Build Decision tree Classification

OUTPUT: 
1. Exploratory analysis of the dataset.
2. Naive bayes output
3. Decision tree output.

EXCEPTION HANDLING: None

```python
#!/usr/bin/env python3
"""Code containing work on breastcancer.csv"""
# Importing the necessary libraries
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import confusion_matrix
from sklearn.metrics import accuracy_score, classification_report
from sklearn.tree import DecisionTreeClassifier

# Import the CSV file using pandas
df = pd.read_csv("breastcancer.csv")

# Exploring the data set
print(df.head())
print("The shape of the dataset is:", df.shape)
# Checking all the coulumn and their types
print("The columns of the dataset is: ",df.columns)
print("The types are",df.dtypes)

# descriptive stats
print(df.describe())

# Dropping the duplicate values from Dataset into test and train
df.drop_duplicates()
print("\n\n\n")
print("\n","Dimensions after removing duplicate data is", df.shape)
# There are no duplicates in the data also

# Dropping the duplicate values from Dataset into test and train
print(df.corr())
plt.figure(figsize=(20,5))
sns.heatmap(df.corr(), annot=True, fmt=".2")

print("\n\n\n\n\n\n")
print("Splitting the dataset")
# Splitting the data into train and test data
X = df.iloc[:,:-1]
y = df.iloc[:,-1:]
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=0, test_size=0.3)
print("Shape of train is", X_train.shape)
print("Shape of test is", X_test.shape)
```

DISCUSSION: From above we get the following outputs
1. Descriptive stats of the dataset.
2. Dataset after removing duplicate values.
3. Correlation matrix.
4. Split data into train and test.

```python
# Applying naive bayes
print("Building the naive bayes")
# y_train = column_or_1d(y_train, warn=True)
model = GaussianNB()
model.fit(X_train, y_train)
classification_t = model.predict(X_test)
mat = confusion_matrix(y_test, classification_t)
labels = ["1=Healthy Controls", "2=Patients"]
sns.heatmap(mat.T, square=True, annot=True, fmt='d', cbar=False,
            xticklabels=labels, yticklabels=labels)
plt.xlabel('true Classification')
plt.ylabel('predicted Classification')
print("Accuracy is ", accuracy_score(y_test, classification_t))
print(classification_report(y_test, classification_t))
```

DISCUSSION: 
1. From above we get the naive bayes output which says our model has accuracy of 57%
2. Out of 20 patients the model has correctly predicted only 11 which is not ideal.

```python
# Dropping the duplicate values from Dataset into test and train
print("Decision Tree:")
model = DecisionTreeClassifier()
model.fit(X_train, y_train)
classification_t = model.predict(X_test)
mat = confusion_matrix(y_test, classification_t)
labels = ["1=Healthy Controls", "2=Patients"]
sns.heatmap(mat.T, square=True, annot=True, fmt='d', cbar=False,
            xticklabels=labels, yticklabels=labels)
plt.xlabel('true Classification')
plt.ylabel('predicted Classification')
print("Accuracy is ", accuracy_score(y_test, classification_t))
print(classification_report(y_test, classification_t))
```

DISCUSSION: 
1. From above we get a decision tree with accuracy of 65% 
2. All of the outputs display the required information.

Q2: The main aim of the data is to predict the motor and total UPDRS scores ('motor_UPDRS' and 
'total_UPDRS') from the 16 voice measures. 
1. Perform exploratory analysis on the data and Remove motor_UPDRS column.
2. Use cross-validation to build a linear regression model to predict total_UPDRS.
3. Use cross-validation to build a regression tree model to predict total_UPDRS.

INPUT: parkinsons_updrs.data

ASSUMPTION: 

DESIGN: 
1. Explore the dataset using different inbuild python functions
2. Divide the dataset into train, validation, test 
3. Handling missing values
4. Handling duplicates
5. build a linear regression model
6. build a regression tree model

OUTPUT: 
1. Exploratory analysis of the dataset.
2. Linear regression output
3. Regression tree model output

EXCEPTION HANDLING: None

```python
#!/usr/bin/env python3
"""Code containing work on parkinsons_updrs.data"""
# Importing the necessary libraries
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import cross_val_score
from sklearn.linear_model import LinearRegression

data = pd.read_csv("parkinsons_updrs.data")
# Exploring the data set
print(data.head())
print("The Dimensions of the dataset is:", data.shape)
# Checking all the coulumn and their types
print("The columns of the dataset is: ",data.columns)
print("The types are", data.dtypes)

# Descriptive stats
print(data.describe())

# Dropping the duplicate values 
data.drop_duplicates()

# Making a correlation matrix
print("Correlation matrix")
print(data.corr())
plt.figure(figsize=(20,5))
sns.heatmap(data.corr(), annot=True, fmt=".2")

print("Splitting the dataset")
# Now we can divide the data into train test
X = ["subject#", "age", "sex", "test_time", "motor_UPDRS", "Jitter(%)", "Jitter(Abs)", "Jitter:RAP", "Jitter:PPQ5", "Jitter:DDP", "Shimmer", "Shimmer(dB)", "Shimmer:APQ3",
"Shimmer:APQ5", "Shimmer:APQ11", "Shimmer:DDA", "NHR", "HNR", "RPDE", "DFA", "PPE"]
y = "total_UPDRS"
X = data[X]
y = data[y]
# print("Dimensions of train data", X.shape)
# print("Dimensions of test data", y.shape)
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=0, test_size=0.3)
print("Shape of train is", X_train.shape)
print("Shape of test is", X_test.shape)
```

DISCUSSION: From above we get following outputs:
1. Descriptive stats of the dataset.
2. Dimensions of the dataset
3. Handling missing values
4. Correlation matrix
5. Splitting into train and test data.

```python
# Linear regression model
print("Building linear Regression Model")

model = LinearRegression()
model.fit(X_train, y_train)
print("Score without using K-Fold Validation", model.score(X_test, y_test))


lin_reg = LinearRegression()
scores = cross_val_score(lin_reg, X_train, y_train, cv=10)
print("Cross-validation scores:", scores)
```

DISCUSSION: From the above linear regression model we have a accuracy of 90.8%

```python
# Building a regression tree model
from sklearn.tree import DecisionTreeRegressor
print("Building linear Regression Model")

model = DecisionTreeRegressor()
model.fit(X_train, y_train)
print("Score without using K-Fold Validation", model.score(X_test, y_test))


DT_reg = DecisionTreeRegressor()
scores = cross_val_score(DT_reg, X_train, y_train, cv=10)
print("Cross-validation scores:", scores)
```

DISCUSSION: 
1. From the above using a decision tree model we have accuracy of 99.8%
2. We see that all the outputs provide the required information.
